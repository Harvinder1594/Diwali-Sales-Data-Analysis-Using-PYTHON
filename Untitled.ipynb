```python
df=pd.read_csv("dd.csv",encoding='unicode_escape')
df.shape
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount']=df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns={'Marital_Status':'Shaadi'})
```

```python
df.columns
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen= df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x="Gender",y="Amount",data=sales_gen)

```

```python
#from above graph we can see that most of the buyers are females and purchasing power of females are greater than men.
```

```python
ax= sns.countplot(data=df,x='Age Group', hue='Gender')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# from above graph we can see that most buyers are of age 26-35 females
```

```python
# Total Amount Vs Age Group
%matplotlib.inline
```

```python
sales_age= df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sales_age.head()
sns.barplot(x='Age Group',y='Amount',data= sales_age)
```

```python
# from above graph we can state that most of the buyer are of the age group 26-35yrs female.
```

```python
STATE
```

```python
# total order from top 10 states
sales_state= df.groupby(["State"],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
# total amount of sales of top 10 states
```

```python
state_am= df.groupby(["State"],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=state_am,x='State',y='Amount')
```

```python
# from above graph we can see that total no. of orders are from U.P,Mahrashtra,Karnatka and max sale state is up then maharashtra then karnatka.
```

```python
Marital-status
```

```python
ax= sns.countplot(data=df,x='Marital_Status')
sns.set(rc={'figure.figsize':(9,5)})

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sale_m= df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Marital_Status',y='Amount',hue='Gender',data=sale_m)
```

```python
# From above graph we can say that most buyers are married women they have high purchasing power.
```

```python
OCCUPATION
```

```python
sns.set(rc={'figure.figsize':(20,5)})

ax=sns.countplot(data=df,x='Occupation')
```

```python
#occupation wise amount spent.
os= df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Occupation',y='Amount',data=os)
```

```python
# from above graph it can be said that most buyers are from IT-sectors.
```

```python
PRODUCT-CATEGORY
```

```python
sns.set(rc={'figure.figsize':(26,5)})
ax= sns.countplot(data=df,x= 'Product_Category')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Product wise Amount spent
pa= df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(['Amount'],ascending=False)
sns.barplot(data=pa,x='Product_Category',y='Amount')
```

```python
# From the above graph we can see that Most of the sold product are from Food,Clothing Apparel.
```
